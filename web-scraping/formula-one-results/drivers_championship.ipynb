```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.formula1.com/en/results.html/2024/drivers.html'
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
headers = soup.find_all('th')
titles = [title.text for title in headers]
titles = titles[1:6]
titles[0] = 'Position'
titles[-1] = 'Points'
```

```python
df = pd.DataFrame(columns=titles)
df
```

```python
table_rows = soup.find_all('tr')
for row in table_rows[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.replace('\n', ' ').strip() for data in row_data]
    individual_row_data[2] = individual_row_data[2][:-4]

    length = len(df)
    df.loc[length] = individual_row_data[1:6]
```
